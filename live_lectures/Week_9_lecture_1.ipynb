## 3D lines

```python
import matplotlib as mpl

import numpy as np

mpl.rcParams['legend.fontsize'] = 10

# define values for x,y,z
theta = np.linspace(-10 * np.pi, 10 * np.pi, 500)
z = np.linspace(-5, 5, 500)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

```


````python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

# define values for x,y,z
theta = np.linspace(-10 * np.pi, 10 * np.pi, 500)
z = np.linspace(-5, 5, 500)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
````

## 3D stem
+ A stem plot in 3D, which plots vertical lines from a baseline to the z-coordinate and places a marker at the tip.

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi)
x = np.cos(theta - np.pi/2)
y = np.sin(theta - np.pi/2)
z = theta

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.stem(x, y, z)

plt.show()

```

````python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi)
x = np.cos(theta - np.pi/2)
y = np.sin(theta - np.pi/2)
z = theta

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.stem(x, y, z)

plt.show()
````

## Making missing value plots
+ Although we can use pandas.isna().sum() or df.info() to examine missing values, it is easier to look at its visualization.

+ A package to visualize the missing value is called missingno
 + missingno provides missing data visualizations and utilities that allows you to get a quick visual summary.
 + Source: https://github.com/ResidentMario/missingno
 + Install: pip install missingno
 + The msno.matrix nullity matrix is a data-dense display to let you quickly visually pick out patterns in data completion.


```python
import missingno as msno
null_pattern = (np.random.random(1000).reshape((50, 20)) > 0.5).astype(bool)
null_pattern = pd.DataFrame(null_pattern).replace({False: None})
msno.matrix(null_pattern)
```

````python
import missingno as msno
null_pattern = (np.random.random(1000).reshape((50, 20)) > 0.5).astype(bool)
null_pattern = pd.DataFrame(null_pattern).replace({False: None})
msno.matrix(null_pattern)
````

````python
null_pattern
````

+ Missing values in white and non-missing values in black.
+ Each row is a line across each column.
+ The sparkline at right summarizes the total non-missing values across all columns for the rows, and points out the rows with the maximum and the minimum number of complete values in the dataset.
+ msno.bar is a simple visualization of nullity by column:

```python
msno.bar(null_pattern)

```

# **Visualization Best Practices**
+ Good visualization is like telling a good story. It should be coherent, clear, and interesting
+ Several aspects are important to make good visualization
 + Data presentation
 + Color choice
 + Text/Annotation
 + Axes
 + Labels
+ With these aspects, some best practice guidelines include:
 + Present data properly
    + Scatter plots are for representing variable relationships
 + Use color sensibly
    + 5-10% of the population has some form of color blindness, red-green color blindness is common
 + Avoid chart junk – adding in extra components (color gradients, 3D effects, graphic background, shading, etc.) to charts
    + Helps black-white printing, and for colorblind
 + Clearly label axes
 + Tailor your visualization to the audience

````python
iris=sns.load_dataset('iris')
planets=sns.load_dataset('planets')
car_crashes=sns.load_dataset('car_crashes')
````

````python
def visualization_report(dataset, save=False):
  plt.figure(figsize=(16,8))
  dataset.hist(bins=50)
  print("Histogram\n")
  if save:
    plt.savefig("Histogram")
  plt.show()

  print("PairPlot\n")
  sns.pairplot(dataset)
  if save:
    plt.savefig("PairPlot")
  plt.show()

  print("Correlation Heatmap\n")
  sns.heatmap(dataset.corr(), annot=True)
  plt.tight_layout()  # auto-adjust margins
  if save:
    plt.savefig("Corr_heatmap")
  plt.show()
````

````python
visualization_report(iris, True)
````

````python
visualization_report(planets)
````

````python
visualization_report(car_crashes)
````
