# Improve processing speed with numexpr
When handling complex expressions, NumPy stores intermediate results in memory. A package called <font color='blue'>numexpr</font> optimizes the compiles array expression on the fly. **It works well for large and complex array operations.**
We will use <font color='blue'>numexpr.evaluate</font> method which takes a string containing an array expression as its first argument. Let us how fast it is


````python
import numpy as np
import numexpr as ne

def distance_matrix_numpy(n):
  r=np.random.rand(n,2)
  r_i=r[:, np.newaxis]
  r_j=r[np.newaxis, :]
  #print(r.shape)
  #print(r_i.shape)
  #print(r_j.shape)
  d_ij=r_j-r_i
  return np.sqrt((d_ij**2).sum(axis=2))
````

````python
%timeit distance_matrix_numpy(10000)
````

````python
def distance_matrix_numexpr(n):
  r=np.random.rand(n,2)
  r_i=r[:, np.newaxis]
  r_j=r[np.newaxis, :]
  d_ij=ne.evaluate("sum((r_j-r_i)**2, 2)")
  return ne.evaluate("sqrt(d_ij)")
````

````python
%timeit distance_matrix_numexpr(10000)
````

## **Grouping**
+ Use Pandas **groupby()**  method to group the underlying data on one or more columns and estimate useful statistics such as count, mean, median, std, min, max, etc.
+ Using pandas groupby size()
 + Group the dataframe on the column “Team” and get the number of rows in each group using the groupby **size()** or **count()** function.
   + the count() function will NOT count any NaN values it encounters.

```python
df = pd.DataFrame({
    'Team' : ['A', 'A', 'B', 'B', 'B', 'C'],
    'Points': [100, 80, 60, 70, 95, 83]
})
# size of each group
print(df.groupby('Team').size())
#or
print(df['Team'].counts())
```

[ref](https://pandas.pydata.org/docs/reference/groupby.html)

+ Groupby mean()

```python
print(df.groupby('Team').mean())
```

+ Groupby max
```python
print(df.groupby('Team').max())
```

+ Groupby median
```python
print(df.groupby('Team').median())
```

## **Aggregations**
+ Pandas agg() method aggregates dataframe using one or more operations over the specified axis.
+ Apply statistics: mean, median, std, min, max, etc.

```python
df = pd.DataFrame([[1, 2, 3],
                    [4, 5, 6],
                    [7, 8, 9],
                    [np.nan, np.nan, np.nan]],
                    columns=['A', 'B', 'C'])


df.agg(['sum', 'min', 'std', 'mean'])

```


+ Aggregate different functions over the columns and rename the index of the resulting DataFrame.

```python
df.agg(max=('A', max), min=('B', min), mean=('C', np.mean))
```
