## **Access Modifiers in Python**

+ Access modifiers limit access to the variables and methods of a class.

+ Python provides three types of access modifiers private, public, and protected.
 + Public Member: Accessible anywhere from outside class.
 + Private Member: Accessible within the class
 + Protected Member: Accessible within the class and its sub-classes

# **Private Members of the Parent Class**
+ We can make some instance variables of the parent class private, which will not be available to the child class.
+ An instance variable can be made private by adding double underscores (__) before its name.

```python
class parent(object):
    def __init__(self):
        self.name = 'Mike'  
        # age is a private instance variable
        self.__age = 42
class child(parent):
    def __init__(self):
        self.year = 1995
        parent.__init__(self)
```

````python
class Parent(object):
    def __init__(self):
        self.name = 'Mike'
        # age is a private instance variable
        self._age = 42
class Child(Parent):
    def __init__(self):
        self.year = 1995
        Parent.__init__(self)
````

````python
My_obj = Child()
# produces an error as ‘age’ is a private instance variable
print(My_obj._age)
````

# **Getters and Setters in Python**
+ Use the getter method to access data members and the setter method to modify the data members.
+ The getters and setters methods are often used to avoid direct access to private variables.


````python
class Student:
    def __init__(self, name, age):
        #public member
        self.name=name
        #private member
        self.__age=age

    def get_age(self):
        return self.__age
````

````python
stud=Student("Sam", 18)
````

````python
print("Age: ",stud.get_age() )
````

# **Protected Member**
+ Protected members are accessible within the class and also available to its sub-classes.
+ It can be defined as adding a single underscore _. to the variable name.

````python
class Company:
    def __init__(self):
        self._project="NLP"

class Employee(Company):
    def __init__(self, name):
        self.name=name
        Company.__init__(self)

    def show(self):
        print("Employee name: ", self.name)
        print("Working on project: ", self._project)
````

````python
emp=Employee("Jessa")
emp.show()
````

````python
print('Project', emp._project)
````

# **Polymorphism**

+ Polymorphism in Python is the ability of an object to take many forms/roles.
+ In polymorphism, a method can process objects differently depending on the class type or data type.
+ Example:
 + Polymorphism in Built-in function <font color='blue'>len()</font>



````python
# len() being used for a string
print(len("Michigan Tech"))
````

````python
# len() being used for a list
print(len([10, 20, 30]))
````

````python
print(len(["Michigan", "Michigan", "30", "Tech"][1]))
````

The built-in function <font color='blue'>len()</font> calculates the length of an object depending upon its type.
+ If an object is a string, it returns the count of characters;  
+ if an object is a list, it returns the count of items in a list.


# **Polymorphism With Inheritance**
+ Using method overriding polymorphism allows us to define methods in a child class that has the same name as the methods in the parent class.
+ The process of re-implementing the inherited method in the child class is known as Method Overriding.
+ Advantages:
 + Effective in extending the functionality by altering the inherited method.
 + When a parent class has multiple child classes, and one of the child classes wants to redefine the method.


+ In this example, we have a vehicle class as a parent and a ‘Car’ and ‘Truck’ as its sub-class, but each vehicle can have different gears, speed, etc.,


````python
class Vehicle:
    def __init__(self, name, color, price):
        self.name = name
        self.color = color
        self.price = price
    def show(self):
        print('Details:', self.name, self.color, self.price)
    def max_speed(self):
        print('Vehicle max speed is 150')
    def change_gear(self):
        print('Vehicle change 6 gear')
````

````python
# inherit from vehicle class
class Car(Vehicle):
    def max_speed(self):
        print('Car max speed is 240')
    def change_gear(self):
        print('Car change 7 gear')
````

````python
# Car Object
car = Car('Car x1', 'Red', 20000)
car.show()
# calls methods from Car class
car.max_speed()
car.change_gear()

````

````python
# Vehicle Object
vehicle = Vehicle('Truck x1', 'white', 75000)
vehicle.show()
# calls method from a Vehicle class
vehicle.max_speed()
vehicle.change_gear()
````

**In polymorphism, Python first checks the object’s class type and executes the corresponding method.**


# **Polymorphism In Class methods**
+ Polymorphism with class methods is useful when we group different objects having the same method.  
+ The same methods can be added to a list or a tuple, and no need to check the object type before calling their methods.


````python
class Ferrari:
    def fuel_type(self):
        print("Petrol")

    def max_speed(self):
        print("Max speed 350")

class BMW:
    def fuel_type(self):
        print("Diesel")

    def max_speed(self):
        print("Max speed is 240")
````

````python
ferrari = Ferrari()
bmw = BMW()
````

````python
# iterate objects of same type
for car in (ferrari, bmw):
    # call methods without checking class of object
    car.fuel_type()
    car.max_speed()
````
