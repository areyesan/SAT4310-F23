# Improve processing speed with numexpr
When handling complex expressions, NumPy stores intermediate results in memory. A package called <font color='blue'>numexpr</font> optimizes the compiles array expression on the fly. **It works well for large and complex array operations.**
We will use <font color='blue'>numexpr.evaluate</font> method which takes a string containing an array expression as its first argument. Let us how fast it is


````python
import numpy as np
import numexpr as ne

def distance_matrix_numpy(n):
  r=np.random.rand(n,2)
  r_i=r[:, np.newaxis]
  r_j=r[np.newaxis, :]
  #print(r.shape)
  #print(r_i.shape)
  #print(r_j.shape)
  d_ij=r_j-r_i
  return np.sqrt((d_ij**2).sum(axis=2))
````

````python
%timeit distance_matrix_numpy(10000)
````

````python
def distance_matrix_numexpr(n):
  r=np.random.rand(n,2)
  r_i=r[:, np.newaxis]
  r_j=r[np.newaxis, :]
  d_ij=ne.evaluate("sum((r_j-r_i)**2, 2)")
  return ne.evaluate("sqrt(d_ij)")
````

````python
%timeit distance_matrix_numexpr(10000)
````

## Let's count the total sales per city

````python
## groupby and sum
city_sales = df.groupby('city')['sales'].apply(sum).rename('city_total_sales').reset_index()
city_sales
````

````python
## The Pandas sum equivalent
city_sales = df.groupby('city')['sales'].sum().rename('city_total_sales').reset_index()
city_sales
````

+  ***We have data that each city has multiple restaurants with sales. We would like to know “What is the percentage of sales each restaurant represents in the city”. ***


+ we need to calculate a city_total_sales and combine it back into the data in order to get the percentage.

+ We can use groupby followed by a transform(sum) returns the same output.


````python
df
````

````python
df['city_total_sales'] = df.groupby('city')['sales'].transform('sum')
df
````

````python
df['pct']=df['sales']/df['city_total_sales']
df['pct']=df['pct'].apply(lambda x: format(x,'.2%'))
df
````
