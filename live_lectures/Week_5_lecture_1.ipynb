````python
#Data structure
import numpy as np

x=np.arange(20, dtype=np.float32)
print(x)
print(x.shape)
print(x.dtype)
print(x.strides)

````

````python
#Indexing (view)
x_view=x.view()
print(x_view)
print(x)
x_view[2]=-1*x_view[2]
print(x_view)
print(x)
````

````python
#Indexing (copy)
x_copy=x.copy()
print(x_copy)
print(x)
x_copy[1]=-1*x_copy[1]
print(x_copy)
print(x)
````

````python
#Vectorization
a=np.arange(10)
b=np.ones(10)
print(a)
print(b)
c=a+b
print(c)
````

````python
#Broadcasting

a=np.arange(10).reshape(2,-1)
print(a)
b=np.ones(2).reshape(2,1)
print(b)
#print(a)
#print(b)
c=a+b
print(c)
````

# **Lists vs Arrays**

````python
import sys
import numpy as np

# Create a Python list
py_list = list(range(900))

# Create a NumPy array
np_array = np.arange(900)

print("Memory size of Python list: ", sys.getsizeof(py_list))
print("Memory size of NumPy array: ", sys.getsizeof(np_array))#np_array.nbytes)
````

````python
import numpy as np
import sys

N = int(1e7)

narray = np.zeros(N);
mylist = []

for i in range(N):
    mylist.append(narray[i])

print("size of np.array:", sys.getsizeof(narray))
print("size of list    :", sys.getsizeof(mylist))
````

# Accessing array  - Updating
+ In numpy, if we take a slide of the original array and change one of its values in a new array, then the original value will be updated as well.  Indexing with slicing syntax is fast because it does not produce a copy of the array.
```python
A =np.arange(6).reshape(2,3)
B = A[1:]
B[:] = [10,20,30]
print(A)
```

````python
A =np.arange(6).reshape(2,3)
B = A#A.copy()
print(A)
print(B)
````

````python
B[1] = [10,20,30]
print(A)
print(B)
````
