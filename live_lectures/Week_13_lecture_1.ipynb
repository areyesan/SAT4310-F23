# **urllib.robotparser**

```python
# importing robot parser class
import urllib.robotparser as rb

bot = rb.RobotFileParser()
bot.set_url('https://www.geeksforgeeks.org//robots.txt')
bot.read()
# check if we can crawl the main site
y = bot.can_fetch('*', 'https://www.geeksforgeeks.org/wp-admin/')
print(y)
```

```python
# importing robot parser class
import urllib.robotparser as rb

bot = rb.RobotFileParser()
bot.set_url('https://www.mtu.edu//robots.txt')
bot.read()
# check if we can crawl the main site
y = bot.can_fetch('*', 'https://www.mtu.edu/carnival/2*/images/')
print(y)
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup)
```

# Web Scraping Demo

```python
#import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
```

```python
page=urllib.request.urlopen("https://forecast.weather.gov/MapClick.php?lat=47.117489900000066&lon=-88.57245689999996#.Y4Po33ZOmbg")
print(page)
```

```python
soup=BeautifulSoup(page)

print(soup)
```

```python
seven_day = soup.find(id="seven-day-forecast")

print(seven_day.prettify())
```

```python
forecast_days = seven_day.find_all(class_="tombstone-container")

len(forecast_days)
```

```python
Tonight = forecast_days[2]
print(Tonight.prettify())
```

```python
period=Tonight.find(class_="period-name").get_text()
short_desc=Tonight.find(class_="short-desc").get_text()
temp=Tonight.find(class_="temp").get_text()

print(period)
print(short_desc)
print(temp)
```

```python
img = Tonight.find("img")
desc = img['title']
print(desc)
```

```python
periods=[pt.get_text() for pt in seven_day.select(".tombstone-container .period-name")]
short_descs=[sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps= [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs=[des["title"] for des in seven_day.select(".tombstone-container img")]
```

```python
print(periods)
print(short_descs)
print(temps)
print(descs)
```

```python
print(len(periods))
print(len(short_descs))
print(len(temps))
print(len(descs))
```

```python
import pandas as pd
weather = pd.DataFrame({
    "period": periods,
    "short_desc": short_descs,
    #"temp": temps,
    "desc":descs
})
weather
```

```python
weather.to_csv("weather_class.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=47.117489900000066&lon=-88.57245689999996#.Y4Po33ZOmbg")
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")

forecast_days = seven_day.find_all(class_="tombstone-container")

Tonight = forecast_days[1]
print(Tonight.prettify())
```

```python
period=Tonight.find(class_="period-name").get_text()
short_desc=Tonight.find(class_="short-desc").get_text()
temp=Tonight.find(class_="temp").get_text()

print(period)
print(short_desc)
print(temp)
```

```python
img = Tonight.find("img")
desc = img['title']
print(desc)
```
