## Heapq â€” Heap queue algorithm
+ The heap queue algorithm, also known as the priority queue algorithm.
+ Heaps are binary trees for which every parent node has a value less than or equal to any of its children. This implementation uses arrays for which heap[k] <= heap[2*k+1] and heap[k] <= heap[2*k+2] for all k, counting elements from zero. For the sake of comparison, non-existing elements are considered to be infinite. The interesting property of a heap is that its smallest element is always the root, heap[0].
+ It is a data structure designed to quickly find and extract the min value from a collection.
 + Uses the heap queue algorithm, also known as the priority queue algorithm.
 + Allows for insertion and extraction of max value
 + Time complexity: O(log(N))
+ import:
   
 + import heapq
 + heapq.heapify(x): Transform list x into a heap in linear time.
 + heapq.heappop(heap): pop and return the smallest item from the heap, maintaining the heap invariant.
 + heapq.heappush(heap, item): push the value item onto the heap, maintaining the heap invariant.

````python
import heapq
df = [10,3,5,8,9,12,20]

df
````

# Efficient iteration with comprehensions and generators

## List comprehensions and generators are faster than using for-loop, as seen in the following example.




````python
def loop():
    res = []
    for i in range(1_000_000):
        res.append(i * i)
    return sum(res)

def comprehension():
    return sum([i * i for i in range(1_000_000)])

def generator():
    return sum(i * i for i in range(1_000_000))
````

````python
%timeit loop()
````

````python
%timeit comprehension()
````

````python
%timeit generator()
````
